```python
#1

# Setting the population restrictions.
population = 10 * (10 ** 6)
tolerance = 1 * (10 ** 6)
population_min, population_max = population - tolerance, population + tolerance

# Selecting the country codes and their respective name that fulfills the population requirements.
country_codes = data[(data["Population"] > population_min) & (data["Population"] < population_max)].index

# Printing the results
for i, _ in enumerate(country_codes):
    print(country_codes[i], end=" - ")
    name = wdi[wdi["Country Code"] == country_codes[i]]["Country Name"].unique()
    print(name[0])
    
```

```python
#2

# Filtering the data
filtered_data = data[data.index.isin(country_codes)]

# Calculating the mean and the standard deviation
GDP_avg = filtered_data["GDP"].mean()
GDP_std = filtered_data["GDP"].std()

# Print the results
print(f"The average is {GDP_avg:.2f} and the standard deviation is {GDP_std:.2f}")
```

```python
#3

# Computing the GDP per capita for each entry in the data data frame
GDP_per_capita = data["GDP"] / data["Population"]

# Filtering the data
filtered_data = GDP_per_capita[GDP_per_capita.index.isin(country_codes)]

# Calculating the mean and the standard deviation
GDP_per_capita_avg = filtered_data.mean()
GDP_per_capita_std = filtered_data.std()

# Printing the results
print(f"The average is {GDP_per_capita_avg:.2f} and the standard deviation is {GDP_per_capita_std:.2f}")
```

```python
#4

# Adding a column with the GDP per capita data
data["GDP per capita"] = GDP_per_capita

# Categorizing the countrys based on the population
intervals = [0, 1000000, 5000000, 15000000, 30000000, 100000000, float('inf')]
intervals_label = ['Tiny', 'Very Small', 'Small', 'Medium', 'Large', 'Huge']
data["Country Size"] = pd.cut(data['Population'], bins=intervals, labels=intervals_label)

# Grouping the data based on the country size
grouped_df_avg = data.groupby("Country Size")["GDP per capita"].mean()
grouped_df_std = data.groupby("Country Size")["GDP per capita"].std()

# Printing the results
result_df = pd.concat([grouped_df_avg, grouped_df_std], axis=1)
result_df.columns = ['Average GDP per capita', 'Std Deviation GDP per capita']
result_df
```

```python
#5

# Defining the function
def above_mean(df, column):
    column_avg = df[column].mean()
    # The lambda function returns 1 if the value is above the mean and is not NaN else it returns a 0 just if the value is not NaN and NaN if the value is NaN
    above_mean_series = df[column].apply(lambda value: 1 if not pd.isna(value) and value > column_avg else 0 if not pd.isna(value) else np.nan)
    return above_mean_series

# Testing the function - So we can make sure the function is working correctly and we can trust the results in the next cells
test_data = pd.DataFrame({'Test': [30, 40, 50, 60, None, 70, 80]})
expected_result = pd.Series([0, 0, 0, 1, np.nan, 1, 1])
result = above_mean(test_data, "Test")
if result.equals(expected_result):
    print("Test passed!")
else:
    print("Test failed!")
```

```python
#6

# Creating a new column with 0's and 1's if the hc value is below or above the mean
data["hc above average"] = above_mean(data, "hc")

# Computing the mean and the standard deviation
hc_mean = data.groupby(["Country Size", "hc above average"])["GDP per capita"].mean()
hc_std = data.groupby(["Country Size", "hc above average"])["GDP per capita"].std()

# Printing the results
result_df = pd.concat([hc_mean, hc_std], axis=1)
result_df.columns = ['Average GDP per capita', 'Std Deviation GDP per capita']
result_df.index = result_df.index.set_levels(result_df.index.levels[1].map({0: "No", 1: "Yes"}), level=1)
result_df
```

```python
#7
data["GINI above average"] = above_mean(data, "GINI")

gdp_avg = data.groupby(["hc above average", "GINI above average"])["GDP per capita"].mean()
gdp_std = data.groupby(["hc above average", "GINI above average"])["GDP per capita"].std()


# Print the results
result_df = pd.concat([gdp_avg, gdp_std], axis=1)
result_df.columns = ['Average GDP per capita', 'Std Deviation GDP per capita']
result_df
```

```python
#8

# Filetring the data
filtered_data = data[(data["hc above average"] == 0) & (data["GINI above average"] == 0)]

# Computing the max value and extracting the respective country code and name
max_value = filtered_data["GDP per capita"].max()
country_code_of_max = filtered_data[filtered_data["GDP per capita"] == max_value].index[0]
country_name_of_max = wdi[wdi["Country Code"] == country_code_of_max]["Country Name"].unique()[0]

# Printing the results
print(f"The country with biggest GDP per capita is {country_name_of_max} ({country_code_of_max}) with a GDP per capita of {max_value:.2f}")
```

```python
#9

# Making the column GINI above average per size
data_per_size = data.groupby("Country Size")["GINI"].mean()

# Here we compute the GINI above average per size column
for label in intervals_label:
    # Step 1: Filter Data by "Country Size" for the current label
    filtered_data = data[data["Country Size"] == label].copy()
    
    # Step 2: Calculate "GINI above average per size" for the filtered data
    filtered_data["GINI above average per size"] = above_mean(filtered_data, "GINI")
    
    # Step 3: Update the "GINI above average per size" values in the original DataFrame "data"
    # Assign the calculated values from filtered_data to the corresponding rows in "data"
    data.loc[data["Country Size"] == label, "GINI above average per size"] = filtered_data["GINI above average per size"].values

# Now is the same thing made in the exercise 8 just switch the columns
filtered_data = data[(data["hc above average"] == 0) & (data["GINI above average per size"] == 0)]
max_value = filtered_data["GDP per capita"].max()
country_code_of_max = filtered_data[filtered_data["GDP per capita"] == max_value].index[0]
country_name_of_max = wdi[wdi["Country Code"] == country_code_of_max]["Country Name"].unique()[0]

print(f"The country with biggest GDP per capita is {country_name_of_max} ({country_code_of_max}) with a GDP per capita of {max_value:.2f}")
```
